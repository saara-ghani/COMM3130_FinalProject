Under the pen name Dr. Seuss, Theodor Seuss Geisel (1904–1991) has become an icon of children’s literature with classics like *Green Eggs and Ham*. Although he began his career in advertising and political cartooning, he is celebrated for his playful rhymes, timeless moral lessons, and unforgettable characters that have captivated generations of readers.

This enduring popularity raises an intriguing question: what makes his books so special and memorable? Is it exclusively his vast creativity in building colorful worlds, or does his writing style set him apart from other children's authors? By examining Dr. Seuss’s rhyming techniques and comparing them to those of his peers, we aim to uncover the key characteristics of Dr. Seuss's style and explore whether his approach truly stands out in the world of children's literature.

### Methods and Findings

**Data Collection**

To draw comparisons in style between Dr. Seuss and other children's authors, we built two corpora - one with Dr. Seuss stories and the other with a compilation of children's stories by a variety of authors. We collected texts from internet sources, copy-pasting texts from found PDFs and transcripts into text files.

Each corpus has 25 works, consisting of both poetry and prose stories, and are targeted towards a 4th grade and below reading level. Both corpora have a close total tokens count, with the Seuss Corpus at 20,653 total tokens and the Comparison Corpus at 19,156 total tokens. Both corpora have a unique token count of around 2,700 tokens. The type-token ratio for the Seuss Corpus is 13.09, and the type-token ratio for the Comparison Corpus is 14.16. This ratio indicates how many unique tokens there are, normalized for how many total tokens there are in the corpus. We see that the Comparison Corpus has a slightly higher type-token ratio, potentially indicating a richer vocabulary or less use of word repetition.






**Global Analysis**

To start off, we analyzed rhyming schemes across entire stories in both corpora. By looking at how many times each word in a story rhymes with any other word within that story, we get a sense of how "rhyme-y" a story is. Because this analysis is based on each individual word in the texts, we can take the rhyme count for each word in each story and look at the rhyme count distribution across the entire corpus. This allows us to examine how "rhyme-y" each corpus is and compare between Seuss and other children's authors.

The histogram below shows a comparison of all rhyming words in the Seuss and Comparison Corpora. We see that the Comparison Corpus has a greater number of words with 0 rhymes compared with the Seuss Corpus. However, as the count of rhymes increases, we see that the Seuss Corpus has higher frequencies of these rhyme count occurrences. This is a good indicator that Dr. Seuss is more "rhyme-y" than the general children's author.

**Local Analysis**

Having determined that there are distinctions between Dr. Seuss's general rhyming habits and the rhyming methods of other children's authors, we narrowed in to examine rhyming practices in the immediate proximity of tokens for each story for both corpora. We created a "chunk" consisting of the 10 words before and the 10 words after each token, or word, in a story. Within the chunk, we examined three elements:
* Word frequency: all other instances of that token
* Total rhyme frequency: instances of a rhyming word to the token, including repetition of words
* Unique rhyme frequency: instances of a rhyming word to the token, excluding repetition of words

We conducted a bootstrap analysis on each of the three categories for both corpora, sampling the data with replacement 10,000 times and calculating the mean difference between the two corpus samples each time. We then calculated the mean of the mean differences as well as the 95% confidence intervals. 

The difference in means for each bootstrapping analysis was greater than zero, indicating the Seuss Corpus has higher frequencies for use of word repetition, total rhyming, and unique rhyming. Our results are statistically significant, as determined for each frequency analysis by a 95% confidence interval that does not contain zero. It is important to note that due to about 100 Seussian words not being present in the pronouncing package that we utilized for rhyme identifications, Dr. Seuss is predicted to be even more rhyme-proficient than our analysis currently shows. Ultimately, this means Dr. Seuss is more rhyme-y overall, using repetition of words as well as a variety of rhymes to develop creative and unique literature!

### Dr. Seuss's Legacy

Beyond the popularity of his books, Dr. Seuss has also had a lasting impact on popular language. From his story, *How the Grinch Stole Christmas!*, published in 1957, we get the word “grinch”. This word began as the name of a Scrooge-like character in Dr. Seuss’s book, but now is used colloquially to mean an unfriendly or mean-spirited person. In fact, the term has reached “real-word” status as it has been defined in the Merriam-Webster dictionary as “a grumpy person who spoils the pleasure of others”. The most evident example of “Seussian” words becoming part of commonly-used English is the word “nerd”. The use of “nerd” has gone through different connotations, having been seen as a heavy-hitting insult prior to the 1980s to describe individuals overly focused on academics or who come across as socially awkward. Now, the word “nerd” has been reclaimed by academically-inclined people and is often used neutrally or endearingly to describe individuals with a passion for or expertise on a particular topic. However, the first appearance of “nerd” in print was in Dr. Seuss’s book, *If I Ran the Zoo*, published in 1950, where a “Nerd” is a type of character found in the zoo. It has been widely speculated that Dr. Seuss created the word “nerd”, but even if that is not true, his use of the word in print helped launched the term into common use.
