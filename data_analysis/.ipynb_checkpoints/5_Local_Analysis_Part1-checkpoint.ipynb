## Comparison Corpus Localized Dataframes

```python
comp_dataframes = []

for x in range(0, len(comparison_files)):
    story_localized = make_localized_rhyme(comparison_files[x], comp_corpus[x], 10)
    story_dataframe = make_localized_dataframe(story_localized)
    comp_dataframes.append(story_dataframe)

for df in comp_dataframes:
    print() # printing an extra line to make it more visibly nice
    display(df)
```

### Average Word and Rhyme Frequency Scores per Story

Average Word Frequency Score by Story in Seuss Corpus

```python
avg_word_score = []
avg_rhyme_score = []
avg_unique_rhyme_score = []
title_list = []
seuss_freq_scores = {}


for x in range(0, len(seuss_files)):
    df = seuss_dataframes[x]
    
    word_freq = round(df['Word Frequency'].sum() / len(df), 1)
    avg_word_score.append(word_freq)
    
    rhyme_freq = round(df['Rhyme Frequency'].sum() / len(df), 1)
    avg_rhyme_score.append(rhyme_freq)

    unique_rhyme_freq = round(df['Unique Rhyme Frequency'].sum() / len(df), 1)
    avg_unique_rhyme_score.append(unique_rhyme_freq)
    
    title_list.append(seuss_files[x])

seuss_freq_scores = {
                     'Title': title_list,
                     'Avg Word Frequency Score': avg_word_score,
                     'Avg Rhyme Frequency Score': avg_rhyme_score,
                     'Avg Unique Rhyme Frequency Score': avg_unique_rhyme_score
}

df1 = pd.DataFrame(seuss_freq_scores)
df1
```

```python
avg_word_score = []
avg_rhyme_score = []
avg_unique_rhyme_score = []
title_list = []
comp_freq_scores = {}


for x in range(0, len(comparison_files)):
    df = comp_dataframes[x]
    
    word_freq = round(df['Word Frequency'].sum() / len(df), 1)
    avg_word_score.append(word_freq)
    
    rhyme_freq = round(df['Rhyme Frequency'].sum() / len(df), 1)
    avg_rhyme_score.append(rhyme_freq)

    unique_rhyme_freq = round(df['Unique Rhyme Frequency'].sum() / len(df), 1)
    avg_unique_rhyme_score.append(unique_rhyme_freq)
    
    title_list.append(comparison_files[x])

comp_freq_scores = {
                    'Title': title_list,
                    'Avg Word Frequency Score': avg_word_score,
                    'Avg Rhyme Frequency Score': avg_rhyme_score,
                    'Avg Unique Rhyme Frequency Score': avg_unique_rhyme_score
}

df2 = pd.DataFrame(comp_freq_scores)
df2
```
